#### Data Visualisation

```python
# Average Scores
female_data = data[data['gender'] == 'female']
avg_female_score = (female_data['math score'].mean() + female_data['writing score'].mean()) / 2

male_data = data[data['gender'] == 'male']
avg_male_score = (male_data['math score'].mean() + male_data['writing score'].mean()) / 2

avg_male_female_score = [avg_male_score, avg_female_score]

# donuts 
plt.pie(avg_male_female_score)
# Add white circle in the centre of the pie chart
circle = plt.Circle((0,0), 0.7, color='white')
# gcf - get current figure
fig = plt.gcf()
## add the white circle
fig.gca().add_artist(circle)
```

### Math Vs Writing Scores based on gender
- On average, males perform better in 'maths' than females
- On average, females perform better in 'writing' than males
- Strong positive correlation between the exam scores
- Girls tend to be good at writibg
- Men tend to be good iwth numbers

```python
sns.pairplot(data, vars=['math score', 'writing score'], hue='gender')
```

### Delve into students performance based on factors

### Income

#### Those on free/reduces meals overall performed worse than those on standard meals
- This could be to economic factors
- Less stability in the house possibly
- f their family receives certain types of public assistance, such as Universal Credit, or if their household income falls below a certain threshold

```python
sns.barplot(x="lunch", y='math score', data=data, color='orange', errorbar=None)
```

#### No correlation between "income" and education

```python
count_data = data.groupby(['parental level of education', 'lunch']).size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.barplot(data=count_data, x='parental level of education', y='count', hue='lunch')
```

#### Suppport the hypothesis
- Education does not equal income
- We can take away that parents education could correlate with student scores

```python
sorted_data = data.groupby('parental level of education')['math score'].mean().reset_index().sort_values(by='math score', ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x="math score", y="parental level of education", data=sorted_data, color='#1f77b4', errorbar=None)
```
