# Student Performance in Exams
#### This analysis helps to understand the influence of the parents background, test preparation etc on students performance

#### Import the essential Libraries

```python
# Import the essential libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import the Dataset

```python
# Import the dataset
data = pd.read_csv('stu_performance.csv')
```

### Data Cleaning

```python
# check the data
data.head()
```

```python
# show information about the data
data.info()
```

```python
# check for missing values
data.isnull().sum()
```

```python
# check for duplicate rows
data.duplicated().sum()
```

```python
# final manual check for outliers
data
```

```python
# add column to find average score across the 3 subjects
data['average score'] = data[['math score', 'writing score', 'reading score']].mean(axis=1)
data.head()
```

### Analysis

```python
# Average Scores
# avg female score
avg_female_score = data[data['gender'] == 'female']['average score'].mean()
# avg male score
avg_male_score = data[data['gender'] == 'male']['average score'].mean()
# avg combines score
avg_male_female_score = [avg_male_score, avg_female_score]

# donuts
plt.title('Exam scores by Gender')
labels = ['Males','Females']
plt.pie(avg_male_female_score, labels=labels, autopct='%1.1f%%')
# Add white circle in the centre of the pie chart
circle = plt.Circle((0,0), 0.7, color='white')
# gcf - get current figure
fig = plt.gcf()
## add the white circle
fig.gca().add_artist(circle)
plt.show()
```

- Scores are measured from all 3 exams
- Strong similarities in performance between both genders

---

```python
sns.pairplot(data, vars=['math score', 'writing score'], hue='gender')
plt.suptitle('Math vs Writing Scores by Gender', y=1.03)
plt.show()
```

- On average, males perform better in 'maths' than females
- On average, females perform better in 'writing' than males
- Strong positive correlation between the exam scores
- Females tend to be good at writing
- Males tend to be good with numbers

---

```python
# histograms
none_score = data[data['test preparation course'] == 'none']['average score']
complete_score = data[data['test preparation course'] == 'completed']['average score']
ind = none_score.index
plt.hist(none_score, color='skyblue',alpha=0.5, label='None', density=True)
plt.hist(complete_score, color='orange',alpha=0.5, label='Completed', density=True)
plt.yticks([])
plt.suptitle('Test preperation Vs Exam scores')
plt.legend()
plt.show()
```

- Those who completed the test prep on average scored higher in the exams

---

```python
sns.barplot(x="lunch", y='average score', data=data, color='orange', errorbar=None)
plt.suptitle('Exam scores Vs Lunch status')
plt.show()
```

- Free/ Reduced is given to students if their family (receives certain types of public assistance, or if their household income falls below a certain threshold)
- Those on free/reduces meals overall performed worse than those on standard meals by 10%
- **Reasons could be:**
- Fewer books, computers, and internet access at home
- Less stability in the house
- Exposure to stress from financial insecurity

---

```python
count_data = data.groupby(['parental level of education', 'lunch']).size().reset_index(name='count')
count_data = count_data.sort_values(by='count', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=count_data, x='parental level of education', y='count', hue='lunch')
plt.suptitle('Parents level of education Vs Lunch status')
plt.show()
```

- Using lunch status as a measurement of income
- No correlation between household income and parents education
- Most children on standard meals, parents only attended some college
- A lot of students whos parents got a masters, are still on free meals

---

```python
sorted_data = data.groupby('parental level of education')['average score'].mean().reset_index().sort_values(by='average score', ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x="average score", y="parental level of education", data=sorted_data, color='#1f77b4', errorbar=None)
plt.suptitle('Parents level of education Vs Exam scores')
plt.show()
```

- There is a strong correlation between the parents level of education and the students exam performance
- Parents with higher education prioritise their childrens education
